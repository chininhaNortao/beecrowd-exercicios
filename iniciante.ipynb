```python
# exercicio 1158

# quantidade de casos
nCasos = int(input())

# cada caso com um inteiro X e um Y
# Y -> soma deles impares consecutivos, a partir de X
# se X for impar, começa por elee
# faz a soma de Y vezes, começando pelo X
# X Y
soma = 0
repeticao = 0

while nCasos > repeticao:
    entrada = input().split()
    x = int(entrada[0])
    y = int(entrada[1])
    
    if x % 2 != 0:
        count_imp = 0
        count = 0
        i = x
        while count_imp != y:
            if i % 2 != 0:
                soma += i
                count_imp += 1
            i += 1
        print(soma)
        count = 0
        soma = 0
        i = 0
    if x % 2 == 0:
        count_imp = 0
        count = 0
        i = x
        while count_imp != y:
            if i % 2 != 0:
                soma += i
                count_imp += 1
            i += 1
        print(soma)
        count = 0
        soma = 0
        i = 0
    repeticao += 1
```

```python
# 1160 - Crescimento Populacional

rep = int(input())
c = 0
while rep > c:
    data = tuple(input().split())
    
    popA, popB, cresA, cresB = data
    
    popA = int(popA)
    popB = int(popB)
    
    cresA = float(cresA)
    cresB = float(cresB)

    cont_anos = 1
    if int(cresB) != 0 or cresB > 0:
        cont_anos = 1
        
    while popB > popA:
        popA += int(popA * (cresA / 100))
        popB += int(popB * (cresB / 100))
        
        if popA > popB:
            break
        # print(f"População A: {popA}")
        # print(f"População B: {popB}")
        # print(f"Ano: {cont_anos}")
        # print()
        cont_anos += 1
        
        
        if cont_anos > 100:
            print("Mais de 1 seculo.")
            break
    
    if cont_anos <= 100:
        print("{} anos.".format(cont_anos))
    c += 1
```

```python
# 1176 - Fibonacci em Vetor
def f(n):
    if n >= 1:
        l = [0,1]
        for i in range(2,n+1):
            l.append(l[i-1] + l[i-2])
        return l[-1]
    elif n == 0:
        return 0

rep = int(input())
c = 0
while rep>c:
    n = int(input())
    
    print(f"Fib({n}) = {f(n)}")
    c += 1
```

```python
# 1789 - A Corrida de Lesmas
def interar(n):
    return int(n)
while True:
    try:
        qtde_l = int(input())
        velocidade = input().split()
        vel_l = list(map(interar,velocidade))
        
        lesma_rapida = vel_l[0]
        for lesma in vel_l:
            if lesma >= lesma_rapida:
                lesma_rapida = lesma
                
        if lesma_rapida < 10:
            print(1)
        elif lesma_rapida >= 10 and lesma_rapida < 20:
            print(2)
        else:
            print(3)
            
    except EOFError:
        break
```
